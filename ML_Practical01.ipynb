```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


df = pd.read_csv('/content/drive/MyDrive/ML Lab1.csv')


X = df[['Area(sq. ft)']]
y = df['Price(lakhs)']


model = LinearRegression()
model.fit(X, y)


slope = model.coef_[0]
intercept = model.intercept_

print("Slope (Coefficient):", slope)
print("Intercept:", intercept)

# predict for example-1000 sq. ft
new_area = [[1000]]
predicted_price = model.predict(new_area)
print(f"Predicted price for 1000 sq. ft: {predicted_price[0]:.2f} lakhs")

```

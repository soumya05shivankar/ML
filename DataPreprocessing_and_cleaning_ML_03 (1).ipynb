To perform missing values handling outlier detection label encoding and feature scaling on student performance or house price dataset


*   identify null values, apply imputation
*   encode categorical data
*   Normalize or standardize numeric features



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
#load dataset
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
print(df.info())
print(df.head())
print("\nMissing Values:\n",df.isnull().sum())
```

```python
df.loc[5:10, 'petal length(cm)'] = np.nan
df.loc[15:18, 'petal width (cm)'] = np.nan
```

```python
df.isnull().sum()
```

```python
#import missing values
df['petal length (cm)'] = df['sepal length (cm)'].fillna(df['sepal length (cm)'].mean())
df['petal width (cm)'] = df['petal width (cm)'].fillna(df['petal width (cm)'].median())
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(df['petal width (cm)'], orient='h')
plt.title('Box Plot')
plt.show()
```
